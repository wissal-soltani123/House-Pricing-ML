```python
# ============================================================
# STEP 6 — LOAD CLEANED DATA
# ============================================================

import pandas as pd

# Load cleaned dataset
df = pd.read_csv("C:/Users/Acer/Desktop/House Pricing R/data/Cleaned/train_clean.csv")
# Check data
print(df.shape)
print(df.head())
```

```python
# ============================================================
# Separate X (features) and y (target)
# ============================================================

# Target variable
y = df["SalePrice"]

# Features (everything except SalePrice and Id)
X = df.drop(columns=["SalePrice", "Id"])

print(X.shape)
print(y.shape)
```

```python
# ============================================================
# STEP 7 — ENCODING
# ============================================================

# Load test dataset
test = pd.read_csv("C:/Users/Acer/Desktop/House Pricing R/data/Cleaned/test_clean.csv")

# Prepare test features
X_test_final = test.drop(columns=["Id"])


# ===============================
# Combine train + test
# ===============================

combined = pd.concat([X, X_test_final], axis=0)

# ------------------------------------------------------------
# This ensures SAME columns after encoding
# ------------------------------------------------------------


# ===============================
# Apply One-Hot Encoding
# ===============================

combined_encoded = pd.get_dummies(combined)

# ------------------------------------------------------------
# Converts categorical variables → numeric
# ------------------------------------------------------------


# ===============================
# Split back
# ===============================

X = combined_encoded.iloc[:len(X), :]
X_test_final = combined_encoded.iloc[len(X):, :]


# ===============================
# Check
# ===============================

                   
print("Train shape after encoding:", X.shape)
print("Test shape after encoding:", X_test_final.shape)
```

```python
# Save encoded datasets
X.to_csv("C:/Users/Acer/Desktop/House Pricing R/data/Cleaned/X_encoded.csv", index=False)
X_test_final.to_csv("C:/Users/Acer/Desktop/House Pricing R/data/Cleaned/X_test_encoded.csv", index=False)
y.to_csv("C:/Users/Acer/Desktop/House Pricing R/data/Cleaned/y.csv", index=False)
```
